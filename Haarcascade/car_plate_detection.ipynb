```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('Images/car-plate.jpg')
```

```python
plt.imshow(img)
```

```python
def display(img):
    fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    return fix_img
```

```python
fix_img = display(img)
plt.imshow(fix_img)
```

```python
plate_classifier = cv2.CascadeClassifier('Images/car-plate.xml')
```

```python
def plate_detect(img):
    plate_rects = plate_classifier.detectMultiScale(img)
    for (x,y,w,z) in plate_rects:
        cv2.rectangle(img,(x,y),(x+w,y+z),(0,255,0),10)
    return img
```

```python
result = plate_detect(fix_img)
plt.imshow(result)
```
