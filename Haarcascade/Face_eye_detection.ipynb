```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
img = cv2.imread('Images/photo.jpg')
```

```python
fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

<h1>Face Classifier</h1>

```python
face_classifier = cv2.CascadeClassifier('Images/face.xml')
```

```python
def isFace(img):
    face_rects = face_classifier.detectMultiScale(img)
    for (x,y,w,z) in face_rects:
        cv2.rectangle(img,(x,y),(x+w,y+z),(255,0,0),10)
    return img
```

```python
result = isFace(fix_img)
plt.imshow(result)
```

<h1>Eye Classifier</h1>

```python
eye_classifier = cv2.CascadeClassifier('Images/eye.xml')
```

```python
fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
def isEye(img):
    eye_rects = eye_classifier.detectMultiScale(img)
    for (x,y,w,z) in eye_rects:
        cv2.rectangle(img,(x,y),(x+w,y+z),(255,0,0),10)
    return img
```

```python
result = isEye(fix_img)
plt.imshow(result)
```

<h1>Full Body Classifier</h1>

```python
fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
face_classifier = cv2.CascadeClassifier('Images/face.xml')
eye_classifier = cv2.CascadeClassifier('Images/eye.xml')
```

```python
faces = face_classifier.detectMultiScale(fix_img,1.3,5)
```

```python
if faces is ():
    print("No Face detected")
```

```python
def face_eye_detect(img):
    face_rects = face_classifier.detectMultiScale(img)
    for (x,y,w,z) in face_rects:
        cv2.rectangle(img,(x,y),(x+w,y+z),(255,0,0),10)
    eye_rects = eye_classifier.detectMultiScale(img)
    for(ix,iy,iw,iz) in eye_rects:
        cv2.rectangle(img,(ix,iy),(ix+iw,iy+iz),(0,0,255),10)
    return img
```

```python
result = face_eye_detect(fix_img)
plt.imshow(result)
```
